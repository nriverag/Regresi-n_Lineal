El objetivo de este cuaderno es presentar el funcionamiento de la regresión lineal en Python y la integración de GitHub en el flujo de trabajo de proyectos de ciencia de datos. Los modelos de regresión lineal son un buen punto de partida para las tareas de modelamiento de datos. Estos modelos son populares porque se pueden ajustar muy rápidamente y son muy interpretables. En términos estadísticos, la regresión lineal o ajuste lineal es un modelo matemático usado para aproximar la relación de dependencia entre una variable dependiente **Y** con un conjunto de variables independientes **X** más un error $\epsilon$ con probabilidad de 1. El [modelo](https://es.wikipedia.org/wiki/Regresi%C3%B3n_lineal) se puede expresar de la siguiente manera: 
$$Y=\beta_{0}+\beta_{1}X_{1}+\dots+\beta_{m}X_{m}+\epsilon$$

La respuesta de una variable de interés conocida como variable de predicción, va a depender de unas variables conocidas como regresoras, predictoras, o variables independientes, típicamente estas variables son cuantitativas, pero también variables cualitativas pueden ser consideradas en el conjunto de las variables predictoras, esto lleva al estudio de modelos de regresión tipo Poisson, Gamma, Beta, etc., según se la naturaleza de los datos.

De manera general, los modelos de regresión tienen los siguientes usos fundamentales: 
1. Descripción y aplicación: Los modelos de regresión son valiosos para describir el tipo de asociación entre la variable respuesta **Y** y las variables predictoras **X**. En este caso, lo que se persigue es resumir las tendencias de los datos y encontrar la forma de asociación entre las variables.
2. Control y predicción: se tiene la posibilidad de predecir el valor que tendría **Y** bajo unas condiciones dadas además es posible encontrar cuáles de los valores de **X** optimizan la variable dependiente **Y**. 
4. Calibración: La regresión es empleada también en problemas de calibración, en donde **Y* es una característica aleatoria, fácil de medir que depende de una variable no aleatoria difícil de medir.

La teoría de regresión es amplia y cubre estimación de parámetros, pruebas de hipótesis, intervalos de confianza, consistencia e insesgamiento, predicción, contraste del comportamiento de los errores y mucho más. Algunos recursospara profundizar en este tema son:

## Breve historia 

Como la mayoría de desarrollos en el campo de la estadística, los aportes a la construcción del cuerpo de conocimiento conocido como regresión lineal o modelos lineales no fue un proceso ordenado. Diferentes investigadores, matemáticos, estadísticos e ingenieros a lo largo del tiempo fueron desarrollando y aplicando nuevas técnicas y conocimientos teóricos. Para conocer esta historia a fondo es recomendable leer los artírculos de [Barnes](https://journals.sagepub.com/doi/pdf/10.1068/a300203), [Gillard](http://mathsdemo.cf.ac.uk/maths/resources/Gillard_Tech_Report.pdf), [Stanton](https://www.tandfonline.com/doi/full/10.1080/10691898.2001.11910537) y [Melo](https://es.scribd.com/document/367470924/modelos-lineales-melo).


En general, las bases teóricas de estos modelos las dieron [Gauss](https://es.wikipedia.org/wiki/Carl_Friedrich_Gauss) y Legendre a comienzos del siglo XIX, ellos describen el método de los mínimos cuadrados desarrollado bajo el supuesto de la normalidad de los errores. Sin embargo, la metodología basada en la [máxima verosimilitud (MV)](https://es.wikipedia.org/wiki/M%C3%A1xima_verosimilitud) propuesta por Fisher en el primer cuarto del siglo XX, da credibilidad a la teoría de los mínimos cuadrados. [Fisher](https://es.wikipedia.org/wiki/Ronald_Fisher) presentó inicialmente el procedimiento de M.V. en 1912 y se considera uno de los desarrollos más importantes del siglo XX en la estadística.

Fisher en 1912 propusó el "criterio absoluto" que fue derivado inicialmente del principio "de la probabilidad inversa". En 1921 desarrolló el principio del "óptimo" el cual se asociaba con la noción de "verosimilitud". El método de MV propuesto
en 1922 produce estimadores que satisfacen los criterios de sufciencia y eficiencia o cuando se hace MV se llega a un mundo de ideas y nomenclatura estadística, tales como parámetros, estadística, verosimilitud, suficiencia, consistencia, eficiencia e información. El método de MV constituye una herramienta importante en el desarrollo de estimación tanto en los modelos lineales clásicos como en la estructura de los modelos lineales generalizados (MLG) y los modelos de dispersión, y en general, de casi todas las áreas del conocimiento científico.

Después de la segunda guerra mundial, se desarrollaron los computadores de alta velocidad y con éstos vino un amplio desarrollo de los modelos lineales, basados principalmente en procedimientos numéricos y gráficos.

Con respecto a su aplicación en la economía, históricamente la aplicación de muestras estadísticas a datos económicos se da por primera vez en 1699 por [Charles Davemant](https://en.wikipedia.org/wiki/Charles_Davenant), que fue un programa de demanda empírica, 1907 Rodolfo Emiri realiza estudios de demanda, pero es a partir de 1930 con la conformación de la Sociedad Econométrica y posterior publicación de la revista Econometrics que tiene un impulso importante el uso de métodos estadísticos y matemáticos dentro de la economía, desarrollo que toma gran auge después de mitad de siglo y con la aparición del computador. El intento de un econometrista es describir el mundo (fenómeno), y una forma de aproximarse a esta descripción es mediante un modelo.

## Planteamiento del algortimo

El objetivo de esta sección es presentar de manera general el procedimiento para realizar análisis de regresión en Python. Se empieza por importar las librerias usuales: 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
```

El primer paso es ajustar un modelo de regresión lineal simple de la forma:
    $$y = ax + b$$
donde $a$ es la pendiente y $b$ es el término de intercepto. 

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

Se usa el comando de ``LinearRegression`` de la libreria de Scikit-Learn

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
```

La pendiente y el intercepto están contenidos en el objeto de modelo. Los parámetros son ``coef_`` y ``intercept_``:

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

Al igual que programas como R, python puede ajustar modelos de regresión lineal múltiple de la forma: 
$$y = a_0 + a_1 x_1 + a_2 x_2 + \cdots+a_k x_k$$
El proceso es similar: 

```python
rng = np.random.RandomState(1)
X = 10 * rng.rand(100, 3)
y = 0.5 + np.dot(X, [1.5, -2., 1.])

model.fit(X, y)
print("Model intercept:", model.intercept_)
print("Model coeficients:    ",model.coef_)
```
